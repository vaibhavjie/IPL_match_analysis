# Import libraris

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read IPL Dataset _ CSV file

```python
ipl_df=pd.read_csv("matches.csv")
```

# 1. Basic Data Exploration
## view the first few rows
## get info about columns and data type

```python
ipl_df.head()
ipl_df.info()
```

# 2. Team Performance

```python
team_wins= ipl_df['winner'].value_counts()
team_wins
```

# 3. Player of the match

```python
Player_of_the_match_counts=ipl_df['player_of_match'].value_counts()
Player_of_the_match_counts
```

# 4. Runs Scored Analysis

```python
average_runs=ipl_df['target_runs'].mean()
average_runs
```

# Bar chart of team wins

```python
plt.figure(figsize=(10,6))
sns.countplot(x='winner',data=ipl_df)
plt.xticks(rotation=90)
plt.title('Number of Wins by team')
plt.xlabel=('Team')
plt.ylabel=('Wins')
plt.show()
```

# Histogram of Runs Scored

```python
plt.Figure(figsize=(8,5))
plt.hist(ipl_df['target_runs'],bins=8)
plt.title('Distribution of Runs Scored Per Match')
plt.xlabel=('Run_Scored')
plt.ylabel=('Frequency')
plt.show()
```
